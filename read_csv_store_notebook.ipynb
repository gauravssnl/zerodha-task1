```python
import redis
```

```python
r = redis.StrictRedis()
```

```python
r.ping()
```

```python
for (index, code,name,opn,high,low,close) in data.itertuples():
    print(index, code, name, opn, high, low, close)
    r.rpush("code", code)
    r.rpush("name", name)
    r.rpush("opn", opn)
    r.rpush("high", high)
    r.rpush("low", low)
    r.rpush("close", close)
    
```

```python
r.keys()
```

```python
top_ten_close = sorted(list(float(x) for x in r.lrange("close", 0, -1)), reverse = True)[:10]
```
